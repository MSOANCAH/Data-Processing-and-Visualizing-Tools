```python
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
(X, y) = make_blobs(n_samples=250, n_features=2, centers=2,cluster_std=1.05, random_state=20)
#two x values and y value 0 or 1
```

```python
X = np.c_[np.ones((X.shape[0])), X]
# plot the original data along with our line of best fit

plt.scatter(X[:, 1], X[:, 2], marker="o", c=y)
#plt.plot(X, Y, "r-")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

(X, y) = make_blobs(n_samples=250, n_features=3, centers=2,cluster_std=1.05, random_state=20)
#two x values and y value 0 or 1

```

```python
X = np.c_[np.ones((X.shape[0])), X]
# plot the original data along with our line of best fit

plt.scatter(X[:, 1], X[:, 2],X[:,3], marker="o", c=y)
#plt.plot(X, Y, "r-")

plt.show()
```
