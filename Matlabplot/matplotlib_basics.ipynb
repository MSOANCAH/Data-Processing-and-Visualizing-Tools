```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4]) #y axis values
plt.ylabel("number")
plt.show() #x axis values automatically come based on y
```

```python
plt.plot([1,2,3,4], [1,4,9,16]) # xand y values (x,y)
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16],label="line") #color red
plt.axis([0, 6, 0, 20]) # x range and y range
# Add a legend
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]

plt.plot(x,y, linestyle="dashed",marker="o",color="g",label='linear') # color and linestyle
# Add a legend
plt.legend()

plt.xlabel("Numbers") #x title
plt.ylabel("Squares") #y title
plt.title("Number-Square") #figure title

plt.grid(True) #figure as grid

plt.xlim(0,5)# xlimit values
plt.xticks([1,2,3,4])

plt.ylim(0,20)# ylimit
plt.yticks([1,4,9,16])

plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]

plt.plot(x,y, linestyle="dashed",marker="o",color="g",label='square line') # color and linestyle
# Add a legend
plt.legend()

plt.xlabel("Numbers") #x title
plt.ylabel("Squares / Cubes") #y title
plt.title("Number-Square-Cube") #figure title

#2nd plot
a=[1,2,3,4]
b=[1,8,27,64]

plt.plot(a,b, linestyle="dashed",marker="o",color="b",label='cube line')
plt.legend()

plt.show()
```
